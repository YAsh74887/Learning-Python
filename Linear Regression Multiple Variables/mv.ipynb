```python
# This is a code of multivariant linear regression 
# By using this code you can predit the salary of employee
# on basis of experience, test score and interview score

import pandas as pd
from sklearn import linear_model
from word2number import w2n
```

```python
df = pd.read_csv("hr_dt.csv")
df
```

```python
df.test_score.median()
df.test_score = df.test_score.fillna(df.test_score.median())
df.experience = df.experience.fillna("zero")
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score','interview_score(out of 10)']].values,df['salary($)'].values)

reg.predict([[2,9,12]])
```
